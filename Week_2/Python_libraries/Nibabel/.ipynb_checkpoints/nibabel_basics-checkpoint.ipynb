# Nibabel Basics Notebook
### *Learning how to load and view MRI scans*

## Import Libraries
We need nibabel for MRI files and matplotlib for plotting.

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

## Upload an MRI File (.nii.gz)
Upload any NIfTI file.

```python
import nibabel as nib
from nibabel.testing import data_path
import os

file_path = os.path.join(data_path, "example_nifti2.nii.gz")
```

## Load MRI File
Nibabel loads the 3D MRI volume as a NumPy array.

```python
mri = nib.load(file_path).get_fdata()[:, :, :, 0]
print('MRI Shape:', mri.shape)
```

## Visualize a Middle Slice
MRI is 3D. We view one slice at a time.

```python
slice_index = mri.shape[2] // 2
plt.imshow(mri[:, :, slice_index], cmap='gray')
plt.title(f'Slice {slice_index}')
plt.axis('off')
```

## Extract All Slices
Convert 3D MRI into 2D slices.

```python
slices = [mri[:, :, i] for i in range(mri.shape[2])]
len(slices)
```

## Normalize a Slice
Normalization improves contrast.

```python
norm_slice = (mri[:, :, slice_index] - np.min(mri)) / (np.max(mri) - np.min(mri))
plt.imshow(norm_slice, cmap='gray')
plt.title('Normalized Slice')
plt.axis('off')
```

## **Exercises**
1. Load **another** MRI file and visualize 5 slices.
2. Write a function `plot_slices(volume, indices)` to plot given slice numbers.
3. Create a function to normalize **all** slices and return the new array.


```python
import nibabel as nib
from nibabel.testing import data_path
import os

example_file = os.path.join(data_path, "example4d.nii.gz")
mri2 = nib.load(example_file)
mri2_3d = mri2.get_fdata()[:, :, :, 0]

print(f"Volume:", mri2_3d.shape)
```

```python
slice_indices = [3, 6, 9, 12, 15]

fig, axes = plt.subplots(1, 5, figsize=(15, 4))

for ax, idx in zip(axes, slice_indices):
    ax.imshow(mri2_3d[:, :, idx], cmap="gray")
    ax.set_title(f"Slice {idx}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
def plot_slices(mri, indices):
    """
    Plot selected slices from a 3D MRI volume.

    Parameters:
    mri : numpy.ndarray (X, Y, Z)
    indices : list of int (slice numbers along Z-axis)
    """
    
    n = len(indices)
    fig, axes = plt.subplots(1, n, figsize=(3*n, 4))

    if n == 1:
        axes = [axes]

    for ax, idx in zip(axes, indices):
        ax.imshow(mri[:, :, idx], cmap="gray")
        ax.set_title(f"Slice {idx}")
        ax.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
def normalize_volume(volume):
    """
    Normalize a 3D MRI volume to range [0, 1].

    Parameters:
    volume : numpy.ndarray

    Returns:
    normalized_volume : numpy.ndarray
    """
    v_min = volume.min()
    v_max = volume.max()

    normalized_volume = (volume - v_min) / (v_max - v_min)
    return normalized_volume
```
